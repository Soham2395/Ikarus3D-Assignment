```python
from ultralytics import YOLO
import cv2
import torch
import numpy as np
import matplotlib.pyplot as plt
from torchvision import transforms
from torchvision.models.segmentation import deeplabv3_resnet50, DeepLabV3_ResNet50_Weights

# Load YOLOv8 Model
model = YOLO("yolov8m.pt")

# Load DeepLabV3 Model for segmentation
segmentation_model = deeplabv3_resnet50(weights=DeepLabV3_ResNet50_Weights.DEFAULT)
segmentation_model.eval()

# Define relevant furniture categories
furniture_classes = ['chair', 'couch', 'bed', 'table', 'desk', 'tv']

# Define colors for segmentation (Floor = Green, Walls = Light Blue)
FLOOR_COLOR = (0, 255, 0)  # Green
WALL_COLOR = (173, 216, 230)  # Light Blue

def detect_furniture(image):
    """Detect furniture in an image using YOLOv8."""
    results = model(image)
    detections = []

    for result in results:
        for box in result.boxes:
            class_id = int(box.cls)
            label = model.names[class_id]
            confidence = box.conf.item()

            if label in furniture_classes and confidence > 0.5:
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                detections.append((label, confidence, (x1, y1, x2, y2)))

                # Draw bounding box
                cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(image, f"{label} ({confidence:.2f})", (x1, y1 - 10),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    return image, detections

def segment_floor_and_walls(image):
    """Perform semantic segmentation for floor and walls."""
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    input_tensor = transform(image).unsqueeze(0)

    with torch.no_grad():
        output = segmentation_model(input_tensor)['out']

    segmentation_mask = output.argmax(1).squeeze().numpy()

    # Create color overlay
    color_mask = np.zeros_like(image)

    # Assign colors based on segmentation labels
    color_mask[segmentation_mask == 1] = FLOOR_COLOR  # Floor
    color_mask[segmentation_mask == 2] = WALL_COLOR   # Wall

    # Blend with original image
    segmented_image = cv2.addWeighted(image, 0.7, color_mask, 0.3, 0)

    return segmented_image

def analyze_room(image_path):
    """Run object detection & segmentation, then save the output."""
    image = cv2.imread(image_path)

    # Detect furniture
    detected_image, detections = detect_furniture(image.copy())

    # Perform floor and wall segmentation
    segmented_image = segment_floor_and_walls(detected_image.copy())

    # Save & show result
    output_path = "furniture_segmentation_output.jpg"
    cv2.imwrite(output_path, segmented_image)
    plt.imshow(cv2.cvtColor(segmented_image, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()

    # Print detected furniture
    print("\nDetected Furniture:")
    for label, conf, box in detections:
        print(f"- {label} (Confidence: {conf:.2f})")

    print(f"\n✅ Analysis saved as {output_path}")

if __name__ == "__main__":
    image_path = "image.jpg"  # Replace with your image path
    analyze_room(image_path)
```
